```python
from transformers import BertTokenizer, BertModel

outputs = []

for txt in [text, text2, 
            text3
            ]:
  encoding = tokenizer.encode_plus(txt, add_special_tokens = True,    
                                 truncation = True, padding = "max_length",
                                 return_attention_mask = True, 
                                 return_tensors = "pt"
                                 )

  model = BertModel.from_pretrained("bert-base-uncased")

  # inputs = tokenizer("The capital of France is [MASK].", return_tensors="pt")

  outputs.append(model(**encoding).pooler_output[0])
```

```python
for o in outputs:
  print(o.size())
```

```python
import torch 
x = torch.stack(tuple([o for o in outputs]), 0)
```

```python
x
```

```python
# WEIGHTS

from numpy import random

random.seed(42)
# W_Q = random.randint(3, size=(x.size()[1], x.size()[0]))
# W_K = random.randint(3, size=(x.size()[1], x.size()[0]))
# W_V = random.randint(3, size=(x.size()[1], x.size()[0]))

W_Q = random.rand(x.size()[1], x.size()[0])
W_K = random.rand(x.size()[1], x.size()[0])
W_V = random.rand(x.size()[1], x.size()[0])
```

```python
# transform to tensors

w_key = torch.tensor(W_K, dtype=torch.float32)
w_query = torch.tensor(W_Q, dtype=torch.float32)
w_value = torch.tensor(W_V, dtype=torch.float32)
```

```python
# multiply words vector for weight matrices to obtain key, query and values

keys = x @ w_key
querys = x @ w_query
values = x @ w_value
```

```python
querys
```

```python
keys
```

```python
attn_scores = querys @ keys.T
```

```python
attn_scores
```

```python
from torch.nn.functional import softmax

attn_scores_softmax = softmax(attn_scores, dim=-1)
print(attn_scores_softmax)
```

```python
weighted_values = values[:, None] * attn_scores_softmax.T[:,:, None]
```

```python
outputs = weighted_values.sum()
```
